```python
json_file = "/Users/daniel/dataviz/cs765-dc3-data-and-code/RealisticData/100x15.json"

with open(json_file) as data_file:
    data = json.load(data_file)
    
csv_file = "/Users/daniel/dataviz/cs765-dc3-data-and-code/RealisticData/50x15.csv"
pd.read_csv(csv_file)
```

```python
noAss = len(data['assignments'])

# Student related properties
studentTotals = {}
studentAvgPostLen = {}
studentAvgLate = {}
studentLate = {}
studentScores = {}
studentAvgScores = {}
studentPosts = {}
studentTotalPosts = {}

# Assignment related properties
for student in data['students']:
    studentTotals[student['sortable_name']] = 0
    totalPostLen = 0
    totalPosts = 0
    late = []
    scores = []
    posts = []
    count = 0
    for grade in student['grades']:
        if grade['score'] is None:
            grade['score'] = 0
        studentTotals[student["sortable_name"]] += float(grade['score'])
        scores.append(grade['score'])
        allPosts = [x['length'] for x in grade['posts']]
        totalPostLen += sum(allPosts)
        totalPosts += 1 #len(allPosts)
        late.append(grade['late'])
        posts.append(sum(allPosts))
        count =+ 1
        
    studentTotalPosts[student["sortable_name"]] = totalPostLen
    studentAvgLate[student["sortable_name"]] = late[int(noAss/2)]
    studentLate[student["sortable_name"]] = late
    studentAvgPostLen[student["sortable_name"]] = totalPostLen / totalPosts
    studentAvgScores[student["sortable_name"]] = sum(scores) / len(scores)
    studentScores[student["sortable_name"]] = scores
    studentPosts[student["sortable_name"]] = posts
```

```python
averagePostLength = np.array(list(studentAvgPostLen.values()))

t_data = [go.Histogram(x=averagePostLength)]

py.iplot(t_data, filename='Average Post length')
```

```python
avgScores = np.array(list(studentAvgScores.values()))

t_data = [go.Histogram(x=avgScores)]

py.iplot(t_data, filename='Average Score')
```

```python
avgLate = np.array(list(studentAvgLate.values()))

t_data = [go.Histogram(x=avgLate)]

py.iplot(t_data, filename='Average Late')
```

```python
averagePostLengthNorm = (averagePostLength-min(averagePostLength))/(max(averagePostLength)-min(averagePostLength))
avgScoresNorm = (avgScores-min(avgScores))/(max(avgScores)-min(avgScores))

# Group data together
hist_data = [averagePostLengthNorm, avgScoresNorm]

group_labels = ['Average Post Length Norm', 'Average Score']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)

# Plot!
py.iplot(fig, filename='Distribution')
```
